در این تمرین قصد داریم به تحلیل داده های فیلم های برتر سینمای جهان به معیار imdb بپردازیم. ابتدا دادههای مربوط به فیلم را از فایل csv فراخوانی میکنیم

```python
movies = pd.read_csv("imdb_movie_rank.csv")
movies.head()
```

```python
movies.describe()
```

<div dir = "rtl" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
        در این دیتاست اگر داده ای وجود دارد که مقدار nullدارد سطرش را حذف کنید. 
        <br/>
    </font>
</div>

```python
movies.info()
```

```python
#dropna
movies= movies.loc[~movies.content_rating.isnull()]
```

```python
movies=movies.reset_index()
```

```python
movies.info()
```

```python
movies.reset_index()
```

<div dir = "rtl" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
        فیلم ها را براساس مدت زمان فیلم به صورت نزولی مرتب کنید و پنجاه تای اول را در فایل csv با نام long_time ذخیره کنید 
        <br/>
    </font>
</div>

```python
movies_duration=movies.sort_values(by=['duration'], ascending=False)
movies_duration.head()
movies_duration.to_csv('long_time')
```

<div dir = "rtl" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
         تابعی بنویسید که نام بازیگر را به عنوان آرگومان بگیرد و تعداد فیلم های برتر از این لیست را که بازی کرده محاسبه کند.
        <br/>
    </font>
</div>

```python
movies.columns
```

```python
movies.head()
```

```python
movies['Actor4']=((movies.actors_list.str.findall(r"'(.*?)'")))
movies.groupby(['actors', 'title']).agg(rating=('star_rating', 'max'))
```

<div dir = "rtl" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
         میانگین امتیاز فیلم ها را براساس ژانر نمایش به دست آورید.
        <br/>
    </font>
</div>

```python
df_rate=movies.groupby('genre').agg(mean_rate=('star_rating', 'mean'))
df_rate.sort_values(by='mean_rate', ascending=False)
```

<div dir = "rtl" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
       ستونی به این داده ها اضافه کنید که در صورت اینکه یک فیلم مدت نمایش آن 150 دقیقه و یا بیشتر بود مقدار صفر و در غیر اینصورت 1 در هر سطر این ستون قرار گیرد. دیتافریم حاصل را در یک فایل csv ذخیره کنید.
        <br/>
    </font>
</div>

```python
movies['more_than_150']=movies['duration'].map(lambda x: 0 if x>150 else 1)
movies.to_csv('movies_files')
movies.head()
```

<div dir = "rtl" style="direction:rtl;line-height:300%;">

    </font>

```python
movies.columns
```

```python
movies.rename(columns={'star_rating': 'star rating','content_rating': 'content rating', 'ctors_list':'ctors list' })
```

        داده ها را براساس ژانر گروه بندی و مرتب کنید و از هر ژانر صرفا سه تای برتر را نمایش دهید

```python
movies_agg = movies.loc[:,['genre', 'star_rating']]
movies_sort= movies_agg.groupby(['genre']).apply(lambda x: x.nlargest(3,'star_rating' ))
movies_sort
```

### Crosstabs are used for categorical data, while pivot tables can be used for both categorical and numerical data. Crosstabs are used to analyze the relationship between two categorical variables, while pivot tables can analyze the relationships between multiple variables, both categorical and numerical.
