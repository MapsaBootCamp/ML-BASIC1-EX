```python
movies.sort_values('duration', ascending=False, inplace=True)
movies.iloc[:50, :].to_csv('long_time.csv')
```

         تابعی بنویسید که نام بازیگر را به عنوان آرگومان بگیرد و تعداد فیلم های برتر از این لیست را که بازی کرده محاسبه کند.

```python
movies['actors_list'] = movies['actors_list'].apply(lambda x: ', '.join(x))

def get_movies_count(actor_name):    
    filtered_movies = movies[movies['actors_list'].str.contains(actor_name, case=False, na=False)]
    return filtered_movies.shape[0]

get_movies_count('Morgan freeman')  
```

```python
movies['mean_rating'] = movies.groupby(['genre']).star_rating.transform('mean')
movies
```
