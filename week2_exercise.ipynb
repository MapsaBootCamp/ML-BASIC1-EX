```python
movies = movies.sort_values(by='duration', ascending=False)
top_long_time = movies.head(50)
top_long_time.to_csv('long_time.csv', index=False)
```

         تابعی بنویسید که نام بازیگر را به عنوان آرگومان بگیرد و تعداد فیلم های برتر از این لیست را که بازی کرده محاسبه کند.

```python
def count_movies_by_actor(actor_name):
    return len(movies[movies['actors_list'].str.contains(actor_name, case=False)])
count_movies_by_actor("Tom Hanks")
```

```python
movies.star_rating.sum()/movies.genre.count()

movies.groupby("genre").agg(rating=("star_rating", "mean")).sort_values('rating',ascending=False)
```

```python
movies['more_150_min'] = movies['duration'].apply(lambda x: x>=150)
movies.to_csv('movie_data_final.csv', index=False)
```

```python
pd.crosstab(movies.genre, movies.content_rating)
```

در پانداس برای ساخت جدول متقابل  بین دو متغیر استفاده می شود. این جدول متقابل نشان می دهد که چند بار هر دو متغیر در کنار هم وجود دارند.
در کد بالا، با استفاده از crosstab، جدول متقابلی بین ستون genre و content_rating از داده های فیلم ساخته می شود. در این جدول، هر سطر نشان دهنده یک ژانر فیلم و هر ستون نشان دهنده یک رتبه محتوایی (content rating) است. هر خانه در جدول نشان می دهد که چند فیلم با ژانر و رتبه محتوایی خاص در داده ها وجود دارد.

corr: در جدول ها میزان همبستگی دو پارامتر عددی را نسب به یکدیگر می سنجد که بازه ای بین منفی یک و یک است. میزان همستگی دو پارامتر یک باشد یعنی عملکرد مثبت مستقیم بریکدیگر دارند یعنی اگر یکی از آن ها کاهش یابد پارامتر به همان میزان کاهش میابد و بلعکس. و طبیعتا همبستگی یک پارامتر نسبت به خودش یک میباشد.و اگر میزان همبستگی برابر صفر باشد یعنی با کاهش یا افزایش یکی از طرفین در طرف دیگر هیچ تغییری ایجاد نخواهدا شد.و اگر منفی یک باشد رابطه طرفین به صورت معکوس یا الاکلنگی است یعنی با افزایش یکی از طرفین طرف دیگر کاهش پیدا خواهد کرد.

        با استفاده از متد sample هفتاد و پنج درصد سطرها را در یک متغیر به نام sample_1 بریزید. درنهایت با استفاده از متد drop از دیتافریم اصلی این سطرها را حذف و دیتافریم جدید را در sample_2 بریزید

متد pivot_table در پانداس برای ساخت جدول متقابل (cross-tabulation) بین دو یا بیشتر از متغیرهای داده ای استفاده می شود. با استفاده از این متد، می توانیم مقدار های مختلف داده ها را بر اساس یک یا چند متغیر دیگر خلاصه کنیم و به صورت یک جدول خروجی نمایش دهیم. این متد قابلیت استفاده از توابع مختلف برای محاسبه مقادیر خلاصه شده را دارد و همچنین از پارامترهای مختلفی مانند نوع تجمیع، مقدار پیش فرض و ... پشتیبانی می کند.
