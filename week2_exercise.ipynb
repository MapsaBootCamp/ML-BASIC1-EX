```python
movies.dropna(axis=0, inplace=True)
```

<div dir = "rtl" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
        فیلم ها را براساس مدت زمان فیلم به صورت نزولی مرتب کنید و پنجاه تای اول را در فایل csv با نام long_time ذخیره کنید 
        <br/>
    </font>
</div>

```python
movies_data2 = movies.sort_values('duration', ascending=False, ignore_index=True)
long_time= movies_data2.iloc[:50, 1]
long_time.to_csv("long_time.csv", index=False)
```

<div dir = "rtl" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
         تابعی بنویسید که نام بازیگر را به عنوان آرگومان بگیرد و تعداد فیلم های برتر از این لیست را که بازی کرده محاسبه کند.
        <br/>
    </font>
</div>

```python
movies_rating = movies.sort_values(['star_rating'], ascending=[0])
def get_number_of_records(actor_name):
    name_filter=movies_rating["actors_list"].str.contains(actor_name)
    return len(movies_rating[name_filter])

get_number_of_records("Tim Robbins")
```

<div dir = "rtl" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
         میانگین امتیاز فیلم ها را براساس ژانر نمایش به دست آورید.
        <br/>
    </font>
</div>

```python
sorted_list=movies.groupby("genre")
sorted_list.star_rating.agg("mean")
```

<div dir = "rtl" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
       ستونی به این داده ها اضافه کنید که در صورت اینکه یک فیلم مدت نمایش آن 150 دقیقه و یا بیشتر بود مقدار صفر و در غیر اینصورت 1 در هر سطر این ستون قرار گیرد. دیتافریم حاصل را در یک فایل csv ذخیره کنید.
        <br/>
    </font>
</div>

```python
def row_mapping_function(row):
    row["less than 150 minutes"] = 1  if row.duration < 150 else 0
    return row
    
movies.apply(row_mapping_function, axis="columns").to_csv("is_less_than_150_minutes.csv", index=False)
movies
```

<div dir = "rtl" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
        عنوان ستونها را به نحوی تغییر دهید که در ستون های دارای _(آندرلاین) فاصله قرار گیرد. به عنوان مثال star_rating به star rating تبدیل می شود.
        <br/>
    </font>
</div>

```python
movies.columns = [column.replace("_", " ") for column in movies.columns]
```

<div dir = "rtl" style="direction:rtl;line-height:300%;">

        داده ها را براساس ژانر گروه بندی و مرتب کنید و از هر ژانر صرفا سه تای برتر را نمایش دهید
