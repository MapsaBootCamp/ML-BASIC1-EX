```python
def actor_top_movie(data,name_of_actor:str,number_of_films=3):
    for index , item in movies['actors_list'].items(): 
        movies.loc[index,'actor_appear']=name_of_actor in item
    return movies[movies['actor_appear']].sort_values(['star_rating'],ascending=0)[:number_of_films]
    
    
```

```python
actor_top_movie(data=movies,name_of_actor='Tim Robbins',number_of_films=5)
```
