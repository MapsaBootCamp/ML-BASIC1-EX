```python
def number_of_top_movies(actor_name: str) -> int:
    return movies.actors_list.map(lambda actors_list: True if actor_name in actors_list else False).value_counts()[True]
```

```python
number_of_top_movies("Aamir Khan")
```

```python
def row_mapping_function(row):
    row["less than 150 minutes"] = 1  if row.duration < 150 else 0
    return row
    
movies.apply(row_mapping_function, axis="columns").to_csv("is_less_than_150_minutes.csv", index=False)
```

The crosstab function in the pandas Python library is used to compute a cross-tabulation table. It takes two or more arrays or Series as input and returns a DataFrame that shows the frequency distribution of the variables. The resulting table displays the count of occurrences of each combination of values from the input arrays. It is a useful tool for analyzing categorical data and identifying relationships between variables.

Pandas crosstab function takes two positional arguments. The first one is index which is values to group by in the rows and the second one is columns which is values to group by in the columns. We pass our genre column of movies data frame as index and content rating as columns to this function. We have a frequency table of these factors in the result.

```python
pd.crosstab(movies.genre, movies["content rating"])
```

The resulting correlation matrix is a square DataFrame where each cell represents the correlation coefficient between two columns.

The pivot_table method allows users to create a spreadsheet-style pivot table as a DataFrame. It takes a DataFrame as input and provides a way to summarize and analyze data by grouping and aggregating values based on one or more columns. The pivot_table method allows users to specify the index, columns, and values to be used in the resulting pivot table, as well as the aggregation function to be applied.

```python
pd.pivot_table(movies, values="star rating", index="genre", columns="content rating")
```
