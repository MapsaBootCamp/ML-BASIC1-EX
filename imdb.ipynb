```python
import pandas as pd
import numpy as np
import re
```

#### در این دیتاست اگر داده ای وجود دارد که مقدار nullدارد سطرش را حذف کنید.

```python
movies = pd.read_csv('imdb_movie_rank.csv').dropna()
```

#### فیلم ها را براساس مدت زمان فیلم به صورت نزولی مرتب کنید و پنجاه تای اول را در فایل csv با نام long_time ذخیره کنید

```python
long_time = movies.sort_values('duration', ascending = False).iloc[:50,:]
long_time
```

#### تابعی بنویسید که نام بازیگر را به عنوان آرگومان بگیرد و تعداد فیلم های برتر از این لیست را که بازی کرده محاسبه کند.

```python
def actor_best_movies(name , series):
    final_lst = []
    #Regex
    series.map(lambda x : re.findall(r'u\'(\w*\s\w*)\'', x))
    for lst in series:
        if name in lst:
            final_lst.append(lst)
    df = movies[movies['actors_list'].isin(final_lst)]
    return df.sort_values('star_rating',ignore_index=True,ascending= False)

actor_best_movies('Morgan Freeman',movies['actors_list'])
```

#### میانگین امتیاز فیلم ها را براساس ژانر نمایش به دست آورید.

```python
# movies.groupby(movies['genre'])['star_rating'].mean()

dict_gener_rate = {}
for each_gener in movies['genre']:
    dict_gener_rate[each_gener] = movies[movies['genre'] == each_gener]['star_rating'].mean()


pd.DataFrame(dict_gener_rate.items())
```

#### ستونی به این داده ها اضافه کنید که در صورت اینکه یک فیلم مدت نمایش آن 150 دقیقه و یا بیشتر بود مقدار صفر و در غیر اینصورت 1 در هر سطر این ستون قرار گیرد. دیتافریم حاصل را در یک فایل csv ذخیره کنید.

```python
movies['more than 150 min'] = 1
movies.loc[(movies[movies['duration'] >= 150].index), 'more than 150 min'] = 0
movies.to_csv('todo_imdb.csv')
movies
```

#### عنوان ستونها را به نحوی تغییر دهید که در ستون های دارای _(آندرلاین) فاصله قرار گیرد. به عنوان مثال star_rating به star rating تبدیل می شود.

```python
for each in movies:
    movies.rename(columns={each : each.replace('_',' ')},inplace=True)
movies
```

#### داده ها را براساس ژانر گروه بندی و مرتب کنید و از هر ژانر صرفا سه تای برتر را نمایش دهید

```python
movies.groupby("genre").apply(lambda x: x.nlargest(3,['star rating']))
```
