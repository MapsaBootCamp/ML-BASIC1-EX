# Generate Data

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
X, y = make_regression(n_samples=100, n_features=1, n_informative=1, random_state=0, noise=15) 
y=y.reshape(100,1)
X.shape,y.shape
```

```python
plt.scatter(X,y)
```

# with numpy:

# Adding 1 to matrix

```python
intercept_ones = np.ones((len(X),1)) 
X_b = np.c_[intercept_ones,X]
X_b.shape
```

# Getting the updated œ¥1,œ¥0 for y=œ¥1x+œ¥0
ùêΩ(œ¥)=1/2ùëö(‚àë(‚Ñéœ¥(ùë•(ùëñ))‚àíùë¶(ùëñ))2
</br>
œ¥ùëó=œ¥ùëó‚àíùõº(‚àÇùêΩ(œ¥)/(‚àÇœ¥ùëó))

```python
def gradient_descent(alpha, x, y, numIterations):
    m = x.shape[0] 
    theta = np.ones((2,1))
    x_transpose = x.transpose()
    for iter in range(0, numIterations):
        hypothesis = np.dot(x, theta)
        loss = hypothesis - y
        J = np.sum(loss ** 2) / (2 * m)        
        gradient = np.dot(x_transpose, loss) / m         
        theta = theta - alpha * gradient  
    return theta
```

```python
w=gradient_descent(0.01, X_b, y, 1000)
w
```

# predict

```python
y_pred=np.dot(X,w[1])+w[0]
```

```python
plt.scatter(X, y)
plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color='red') # predicted
plt.show()
```
