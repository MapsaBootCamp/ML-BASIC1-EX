## mathplotlib

#### 1 - read company_sales_data.csv with pandas

#### 2 - make a plot with total_profit as x axis and month_number columns as y axis

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame(pd.read_csv('company_sales_data.csv'))
```

```python
df.head()
```

```python
df.info()
```

```python
df=df.astype({'month_number':'int64'})
```

```python
plt.plot('month_number','total_profit',  data=df)
plt.show()
```

#### 3 - add "profit" as x label and "Month number" as y label  "company profit" as title  


#### 4 - each month_number should show in x axis and y axis should be [100000, 200000, 300000, 400000, 500000]

```python
plt.plot('month_number','total_profit',  data=df)
plt.xlabel('Month number')
plt.ylabel('profit')
plt.title('company profit')
plt.xticks(np.arange(1,13,1))
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

### style change

#### 5 - make your plot appearance like below(5 parameters should change)

```python
plt.plot('month_number','total_profit',  data=df, marker='o',color='red', mec='y',mfc= 'w',linestyle='--')
plt.xlabel('Month number')
plt.ylabel('profit')
plt.title('company profit')
plt.xticks(np.arange(1,13,1))
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

### bar plot 
#### 6 - create plot with toothpaste and shampoo column (if you can make background girdy)
#### use width and align='edge' to print bars beside each others not top of each others)

```python
plt.subplot()
plt.bar('month_number',height='toothpaste', data=df, label='Ftoothpasete sales data', align='edge',width = +0.25)
plt.bar('month_number',height='shampoo', data=df, label='shampoo sales data',align='edge',width = -0.25)
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.title('Ftoothpaste and shampoo sales data')
plt.xticks(np.arange(1,13,1))
plt.grid(linestyle=':')
plt.show()
```

#### 9 - load titanic dataset with seaborn  and plot the count of survivors according to sex

```python
df_titanic=pd.DataFrame(pd.read_csv('train.csv'))
```

```python
df_titanic_sur=df_titanic.groupby(['Survived', 'Sex']).agg(count=('Survived', 'count'))
```

```python
df_titanic_sur=df_titanic_sur.reset_index()
```

```python
sns.barplot(data=df_titanic_sur,x='Survived',y='count', hue='Sex')
```
