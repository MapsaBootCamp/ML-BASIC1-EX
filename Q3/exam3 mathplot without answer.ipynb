### style change

#### 5 - make your plot appearance like below(5 parameters should change)

```python
plot = data['total_profit'].plot(
    yticks = [100000,200000,300000,400000,500000],
    ylabel = 'profit',
    xticks = data['month_number'],
    xlabel = "Month number",
    linestyle = "--",
    color = "red",
    marker = "o",
    linewidth = "3",
    markersize = "5",
    markerfacecolor = "yellow"
)
```
