```python
df = df.sort_values(by = "duration" )
path = '/home/mohe/Desktop/long_time.csv'
df.tail(50).to_csv(path, index = False)
```

```python
def count_actor(actor_name):
    if len(df[df['actors_list'].str.contains(str(actor_name))]) == 0:
        print("this actor isn't play in this movie dataframe")
        return
    return len(df[df['actors_list'].str.contains(str(actor_name))])

count_actor("Russell Crowe")
```
