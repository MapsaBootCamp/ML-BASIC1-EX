### 1- با استفاده از دادگان ارائه شده محاسبه کنید هر مشتری چند درصد از هر محصول را خریده است. (با دقت دهگان درصد)

```python
cpm_p_table = pd.pivot_table(data, values= 'Count', index=['Customer'],
                       columns=['Product', 'Model'], fill_value=0)
customer_total = cpm_p_table.sum(axis=1)
percentage_table = cpm_p_table.div(customer_total, axis=0) * 100

percentage_table
```

### 2- به هر سطر لیست دادگان میزان کل محصول خریداری شده اضافه کنید و بنا به شماره مشتری، محصول و مدل مرتب کنید

```python
df1= pd.pivot_table(data, values= ['Count', 'Volume'], index=['Customer','Product', 'Model'], fill_value=0)
df1
```

```python
sum_of_model_values = df1.groupby(['Customer', 'Product'])['Count'].transform('sum')
sum_of_volume_values = df1.groupby(['Customer', 'Product'])['Volume'].transform('sum')

# Add the new columns to the pivot table
df1['Sum of Model'] = sum_of_model_values
df1['Sum of Volume'] = sum_of_volume_values
df1
```

```python
import pandas as pd

record = {

'Name': ['Ankit', 'Amit', 'Aishwarya', 'Priyanka', 'Priya', 'Shaurya' ],
'Age': [21, 19, 20, 18, 17, 21],
'Stream': ['Math', 'Commerce', 'Science', 'Math', 'Math', 'Science'],
'Percentage': [88, 92, 95, 70, 65, 78] }

# create a dataframe
dataframe = pd.DataFrame(record, columns = ['Name', 'Age', 'Stream', 'Percentage'])

dataframe
```

```python
df = dataframe[dataframe['Age'] >= 18].sort_values(by='Percentage', ascending=False).reset_index(drop=True)
df.index = ['A', 'B', 'C', 'D', 'E'][:len(df)]
df
```
