```python
#TODO


sorted_movies = movies.sort_values(by=['genre', 'star rating'], ascending=False)

top_movies_by_genre = {}

for genre in sorted_movies['genre'].unique():
    top_movies_by_genre[genre] = sorted_movies[sorted_movies['genre'] == genre][:3]

for genre, movies in top_movies_by_genre.items():
    print(f"Genre: {genre}")
    print(movies)
    print('\n')
```
