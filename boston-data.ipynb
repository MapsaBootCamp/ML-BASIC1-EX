میانگین تعداد اتاق در هر خانه
AGE - نسبت واحدهای تحت اشغال ساخته شده قبل از سال 1940

DIS - weighted distances to five Boston employment centres
فاصله های وزنی تا پنج مرکز استخدامی بوستون

RAD - index of accessibility to radial highways
شاخص دسترسی به بزرگراه های شعاعی
مالیات - نرخ مالیات بر دارایی با ارزش کامل به ازای هر 10000 دلار

TAX - full-value property-tax rate per $10,000
نرخ مالیات بر دارایی با ارزش کامل به ازای هر 10000 دلار
PTRATIO - pupil-teacher ratio by town
 نسبت دانش آموز به معلم بر اساس شهر

1000 (Bk - 0.63)^2 که در آن Bk نسبت سیاهپوستان بر اساس شهر است

LSTAT - % lower status of the population
٪ وضعیت پایین تر از جمعیت

MEDV - Median value of owner-occupied homes in $1000's
####میانگین ارزش خانههای تحت اشغال در 1000 دلار

```python
X = df.drop("medv" , inplace = False, axis= 1)
Y = df["medv"]
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test ,y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=45)
```

```python
X_train.info()  
```

```python
train_data= X_train.select_dtypes(include=[np.number])
```

```python
test_data= X_test.select_dtypes(include=[np.number])
```

```python
# def shuffle_and_split_data(df, test_ratio):
#     shuffled_indices = np.random.permutation(len(df))
#     test_set_size = int(len(df) * test_ratio)
#     test_indices = shuffled_indices[:test_set_size]
#     train_indices = shuffled_indices[test_set_size:]
#     return df.iloc[train_indices], df.iloc[test_indices]
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
std_scaler.fit(train_data)
```

```python
X_train_scale = std_scaler.transform(train_data)
X_test_scale = std_scaler.transform(test_data)
```

```python
from sklearn.linear_model import LinearRegression
lr_model= LinearRegression()
lr_model.fit(X_train_scale , y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
y_predict = lr_model.predict(X_test_scale)
y_predict
```

```python
r2_score(y_test, y_predict)
```

```python
mean_squared_error(y_test, y_predict, squared= False)
```

```python
mean_absolute_error(y_test, y_predict)
```

```python
mean_absolute_percentage_error(y_test, y_predict)
```
