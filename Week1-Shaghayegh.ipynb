###### part 6

```python
for time in movies["duration"]:
    if time >= 150:
        movies["150_or_more"]=0
    else:
        movies["150_or_more"]=1
movies
```

```python
pd.crosstab(movies.genre, movies.content_rating)
```

pivot_table() is used to create a spreadsheet-style pivot table as a DataFrame. The levels in the pivot table will be stored in MultiIndex objects (hierarchical indexes) on the index and columns of the result DataFrame.

like in music sites, using this method, we can create a table having the users as the rows and the songs in the columns and the cells in this table are the ratings of different musics given by each user.
