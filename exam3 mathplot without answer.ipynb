## mathplotlib

#### 1 - read company_sales_data.csv with pandas

#### 2 - make a plot with total_profit as y axis and month_number columns as x axis

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('company_sales_data.csv')
```

```python
df.head()
```

```python
plt.plot(df.month_number, df.total_profit)
```

#### 3 - add "profit" as y label and "Month number" as x label  "company profit" as title  


#### 4 - each month_number should show in x axis and y axis should be [100000, 200000, 300000, 400000, 500000]

```python
plt.plot(df.month_number, df.total_profit)
plt.xlabel('Month number')
plt.ylabel('profit')
plt.title('company profit')
plt.xticks(df.month_number)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.grid()
```

### style change

#### 5 - make your plot appearance like below(5 parameters should change)

```python
plt.plot(df.month_number, df.total_profit,color='r',marker='o',markerfacecolor='y',
         linestyle='--',linewidth=3)
plt.xlabel('Month number')
plt.ylabel('profit')
plt.title('company profit')
plt.xticks(df.month_number)
plt.yticks([100000, 200000, 300000, 400000, 500000])
```

### bar plot 
#### 6 - create plot with toothpaste and shampoo column (if you can make background girdy)
#### use width and align='edge' to print bars beside each others not top of each others)

```python
plt.bar(df.month_number, df.toothpaste, width= 0.4, label = 'toothpaste', align='edge')
plt.bar(df.month_number, df.shampoo, width= -0.4, label = 'shampoo', align='edge')
plt.xlabel('Month Number')
plt.legend(loc=0)
plt.xticks(df.month_number)
plt.grid(True, linewidth= 1, linestyle="--")
```
