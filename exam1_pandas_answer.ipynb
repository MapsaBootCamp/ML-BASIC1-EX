```python
import pandas as pd
```

```python
data = pd.read_csv('quize_pandas')
```

```python
data.head(5)
```

```python
cpm=data.pivot_table(index=['Customer'],columns=['Product','Model'],values=['Count','Volume'],aggfunc='sum',fill_value=0)
cpm
```

```python
cpm.apply(lambda x:round(100*(x/x.sum()),ndigits=2)).dropna(axis=1)
```

```python
count_val_data=data.groupby(['Customer','Product','Model']).sum()
count_data=count_val_data.groupby(['Product','Model'])['Count'].sum()
vol_data=count_val_data.groupby(['Product','Model'])['Volume'].sum()
```

```python
final=pd.merge(count_val_data,count_data,on=['Product','Model'],suffixes=('','_total')).merge(vol_data,on=['Product','Model'],suffixes=('','_total'))
# final=pd.merge(final,vol_data,on=['Product','Model'],suffixes=('','_total'))
final.rename(columns={'Count':'Count_old','Volume':'Volume_old'},inplace=True)
final
```
