### 1- با استفاده از دادگان ارائه شده محاسبه کنید هر مشتری چند درصد از هر محصول را خریده است. (با دقت دهگان درصد)

```python
import pandas as pd
data = pd.read_csv('quize_pandas.zip', compression='zip')
data.head()
```

```python
cpm_p_table = pd.pivot_table(data, values=["Count","Volume"], index='Customer', columns=('Product','Model'), aggfunc='sum', fill_value=0)
cpm_p_table.fillna(0,inplace=True)
cpm_p_table=cpm_p_table.apply((lambda x:round(100*(x/x.sum()),2)) ,axis=0)
cpm_p_table.dropna(axis=1,inplace=True)
cpm_p_table
```
