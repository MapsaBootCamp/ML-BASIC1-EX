```python
group = df['Product'].groupby(df['Customer Id']).unique()
list(group)
```

```python
df2=pd.DataFrame(group)
seris=list(df2['Product'])
dfss=pd.Series(data=seris)
dfss
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
te = TransactionEncoder()
te_ary = te.fit(dfss).transform(dfss)
dfg = pd.DataFrame(te_ary, columns=te.columns_)
dfg
```

```python

frequent_itemsets=apriori(dfg, min_support=0.01,use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
frequent_itemsets[ (frequent_itemsets['length'] == 3)].sort_values(by='support',ascending=False)
```

```python
frequent_itemsets[ (frequent_itemsets['length'] == 3)].sort_values(by='support',ascending=False)[0:2]
```
