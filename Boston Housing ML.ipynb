# Feature Selection
### According to the HEATMAP, rad and tax features have 0.9 correlation, that is great. Therefore, one of them can be removes from feature set.

```python
# features = features.drop('tax', axis = 1)
# OR
features = features.drop('rad', axis = 1)
```

# Split test and train sets

```python
from sklearn.model_selection import train_test_split
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.3, random_state=10)
```

# Handling Categorical Attributes

```python
from sklearn.preprocessing import OneHotEncoder
# cat_encoder = OneHotEncoder()
```

```python
def convert_cat(data, feature):
 cat_encoder= OneHotEncoder(sparse_output=False).set_output(transform="pandas")
 one_hot_encoded=cat_encoder.fit_transform(data[[feature]])
 data = pd.concat([data,one_hot_encoded],axis=1).drop(columns=[feature])
 return data
```

## Train Set

```python
features_train = convert_cat(features_train, 'chas') 
```

## Test Set

```python
features_test = convert_cat(features_test, 'chas') 
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

# Training

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(features_train, target_train)
```

```python
linear_model.coef_
```

```python
linear_model.intercept_
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
```

```python
evaluation_house_price = linear_model.predict(features_train)
```

```python
r2_score(target_train, evaluation_house_price)
```

```python
mean_squared_error(target_train, evaluation_house_price, squared=False)
```

```python
mean_absolute_error(target_train, evaluation_house_price)
```

```python
mean_absolute_percentage_error(target_train, evaluation_house_price)
```
