## mathplotlib

#### 1 - read company_sales_data.csv with pandas

#### 2 - make a plot with total_profit as x axis and month_number columns as y axis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_df = pd.read_csv(r'D:\machinelearning\company_sales_data.csv')
```

```python
sales_df.columns
```

```python
# total_profit as x axis and month_number columns as y axis
plt.plot(sales_df['total_profit'], sales_df['month_number'])
plt.xlabel('total_profit')
plt.ylabel('month_number')
plt.show()
```

#### 3 - add "profit" as x label and "Month number" as y label  "company profit" as title  


#### 4 - each month_number should show in x axis and y axis should be [100000, 200000, 300000, 400000, 500000]

```python
#3 - add "profit" as x label and "Month number" as y label "company profit" as title
plt.subplots(figsize=(6, 4))
plt.plot(sales_df['month_number'], sales_df['total_profit'])
plt.xlabel('Month number')
plt.ylabel('profit')
plt.title('company profit')
plt.yticks(np.arange(100000, 500000, step=100000))
plt.show()
```

### style change

#### 5 - make your plot appearance like below(5 parameters should change)

```python
plt.subplots(figsize=(6, 4))
plt.plot(sales_df['month_number'], sales_df['total_profit'], '--r', marker='o', markersize=8, markerfacecolor='yellow', markeredgecolor='red', linewidth=2)
plt.xlabel('Month number')
plt.ylabel('Profit')
plt.title('Company Profit')
plt.yticks(np.arange(100000, 500000, step=100000))
plt.xticks(np.arange(1, 12, step=1))
plt.show()
```

### bar plot 
#### 6 - create plot with toothpaste and shampoo column (if you can make background girdy)
#### use width and align='edge' to print bars beside each others not top of each others)

```python
plt.subplots(figsize=(6, 3))
X_axis = np.arange(1, 13, 1)
plt.xlim(0.5, 12.5)
plt.bar(X_axis +0.1, sales_df['toothpaste'], width=0.2, label = 'Toothpaste sales data')
plt.bar(X_axis -0.1, sales_df['shampoo'], width=0.2, label = 'Shampoo sales data')
plt.xticks(X_axis)
plt.yticks(np.arange(0, 9000, 1000))
plt.xlabel("month number", fontsize=9)
plt.ylabel("sales units in number", fontsize=9)
plt.legend()
plt.grid(linestyle='--')
plt.title('Toothpaste and shampoo sales data')
plt.show()
```

```python
import yfinance as yf
# Define the date range
start_date = "1968-12-01"
end_date = "2020-12-01"

# Fetch DJIA data
djia = yf.download("^DJI", start=start_date, end=end_date)

# Select only the 'Adj Close' column
djia = djia['Adj Close']

# Rename the column to 'Price'
djia = djia.rename('Price')

# Reset the index
djia = djia.reset_index()

# Print the first few rows of the dataframe
print(djia.head())

# Save the data to a CSV file if needed
# djia.to_csv('djia_data.csv', index=False)
```

### 10 - 
use seaborn and make plot with dowjones dataet like below

```python
plt.subplots(figsize=(6, 3))
sns.lineplot(data=djia, x='Date', y='Price', color='blue')
plt.title('Dow Jones Industrial Average (DJIA) from 1914 to 1968')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### what function can we use instead of kind="line" in seaborn? and the output with pervois one are the same

```python
plt.subplots(figsize=(6, 3))
sns.pointplot(data=djia, x='Date', y='Price', color='blue', markers='o')
plt.title('Dow Jones Industrial Average (DJIA) from 1914 to 1968')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
