## mathplotlib

#### 1 - read company_sales_data.csv with pandas

#### 2 - make a plot with total_profit as x axis and month_number columns as y axis

```python
# Question 1:
companysalesdata_df = pd.read_csv('company_sales_data.csv')
companysalesdata_df
```

#### 3 - add "profit" as x label and "Month number" as y label  "company profit" as title  


#### 4 - each month_number should show in x axis and y axis should be [100000, 200000, 300000, 400000, 500000]

```python
# Question 2, 3 and 4:
plt.plot(companysalesdata_df['month_number'], companysalesdata_df['total_profit'])
plt.xlabel('Month number')
plt.ylabel('Profit')
plt.title('Company profit')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

### style change

#### 5 - make your plot appearance like below(5 parameters should change)

```python
# Question 5:
plt.plot(companysalesdata_df['month_number'], companysalesdata_df['total_profit'], '--ro', markerfacecolor='yellow', linewidth=3)
plt.xlabel('Month number')
plt.ylabel('Profit')
plt.title('Company profit')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

### bar plot 
#### 6 - create plot with toothpaste and shampoo column (if you can make background girdy)
#### use width and align='edge' to print bars beside each others not top of each others)

```python
# Question 6:
plt.bar(companysalesdata_df['month_number']-0.1, companysalesdata_df['toothpaste'], width=0.2, label = 'Face Cream sales data', align='edge')
plt.bar(companysalesdata_df['month_number']+0.1, companysalesdata_df['shampoo'], width=0.2, label = 'Face Wash sales data', align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Facewash and facecream sales data')
plt.grid(True, linestyle = '--')
plt.legend()
plt.show()
```

#### 9 - load titanic dataset with seaborn  and plot the count of survivors according to sex

```python
# Question 9:
titanic_df = pd.read_csv('train.csv')

sns.catplot(x ="Survived", hue ="Sex", kind ="count", data = titanic_df)
```

####  10 - plot the below plot from titanic dataset

```python
# Question 10:
X = titanic_df.groupby(['Survived'])['Pclass'].mean()
sns.barplot(x = X, y = titanic_df['Survived'], hue =titanic_df['Survived'], orient = 'h', legend=False)
plt.title('Survivors')
plt.show()
```

#### what function can we use instead of kind="line" in seaborn? and the output with pervois one are the same

```python
dowjones = yf.download('DJI', start='1970-12-01', end='1990-12-01')
dowjones = dowjones.reset_index()
dowjones = dowjones.rename(columns={"Adj Close": "Price"})
plt.figure(figsize=(14, 5), dpi=100)
plt.plot(dowjones['Date'], dowjones['Price'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
