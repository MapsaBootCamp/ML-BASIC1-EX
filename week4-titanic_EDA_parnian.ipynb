```python
def bar_plot(df, column_name):
    """
    Create a bar plot for a given categorical column in the DataFrame.
    """
    # Count the values in the column
    value_counts = df[column_name].value_counts()

    # Plot the bar chart
    plt.figure(figsize=(9, 3))
    value_counts.plot(kind='bar', color='skyblue')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.title(f'Bar Plot for {column_name}')
    plt.show()
```

```python

# Define the list of columns to exclude
names_to_delete = ['PassengerId', 'Ticket', 'Cabin', 'Sex', 'Survived', 'Name']

# Get the list of categorical columns excluding those in names_to_delete
categorical_columns = [col for col in titanic_df.columns if col not in names_to_delete and titanic_df[col].dtype == 'object']

# Create bar plots for categorical columns
for col in categorical_columns:
    bar_plot(titanic_df, col)
```

        <br/>
        <font>
        یک تعریف از داده های پرت به این صورت است که هر داده ای که به اندازه یک و نیم برابر طول جعبه (<a href="https://en.wikipedia.org/wiki/Interquartile_range">IQR</a>) از مقادیر چارک اول کوچکتر و یا از چارک دوم بزرگتر باشد داده پرت محسوب میشود. تابعی بنویسید که با گرفتن لیست ستونها اندیس داده های پرت را در آن ستونها تشخیص دهد. در نهایت این داده های پرت را از دیتافریم اولیه حذف کنید و دیتافریم را آپدیت کنید (راهنمایی:‌از متد drop بر روی دیتافریم استفاده کنید. فقط حواستان باشد که در نهایت متد reset_index را با آرگومان drop=True فراخوانی کنید)
        </font>
    </font>
</div>

```python
def detect_outliers(df,features):
  outlier_indexes = []  # Initialize an empty list to store outlier indexes

  for feature in features:
      q1 = df[feature].quantile(0.25)
      q3 = df[feature].quantile(0.75)
      IQR = q3 - q1
      lower_bound = q1 - 1.5 * IQR
      upper_bound = q3 + 1.5 * IQR

      # Find indexes of rows that are outliers for the current feature
      outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)].index
      outlier_indexes.extend(outliers)

  return list(set(outlier_indexes))
```

```python
droppy = detect_outliers(titanic_df,["Age","SibSp","Parch","Fare"])
```

```python
outliers = titanic_df.loc[detect_outliers(titanic_df,["Age","SibSp","Parch","Fare"])]
```
