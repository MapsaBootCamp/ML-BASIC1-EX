```python
from sklearn.datasets import fetch_california_housing
import numpy as np
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

# Generate data

```python
X, y = make_regression(n_samples=100, n_features=1, n_informative=1, random_state=0, noise=15) 
y=y.reshape(100,1)
X.shape,y.shape
```

```python
plt.scatter(X,y)
```

# Adding 1 to matrix

```python
ones=np.ones((X.shape[0],1))
X_b=np.hstack((ones,X))
```

# Getting w :  w=(XX⊤)−1X⊤y
a=(XX⊤)−1
</br>
b=X⊤y
</br>
w=a.b

```python
a=np.linalg.pinv(np.dot(X_b.T,X_b))
b=np.dot(X_b.T,y)
W=np.dot(a,b)
W
```

# Predict

```python
y_pred=np.dot(X,W[1])+W[0]
```

```python
plt.scatter(X, y)
plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color='red') # predicted
plt.show()
```
