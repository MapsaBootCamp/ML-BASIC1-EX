```python
import pandas as pd

record = {

'Name': ['Ankit', 'Amit', 'Aishwarya', 'Priyanka', 'Priya', 'Shaurya' ],
'Age': [21, 19, 20, 18, 17, 21],
'Stream': ['Math', 'Commerce', 'Science', 'Math', 'Math', 'Science'],
'Percentage': [88, 92, 95, 70, 65, 78] }

# create a dataframe
dataframe = pd.DataFrame(record, columns = ['Name', 'Age', 'Stream', 'Percentage'])

dataframe
```

```python
dataframe.rename(columns={'Name': 'Firstname','Percentage':'Percentage_sorted'}, inplace=True)
dataframe
```

```python
dataframe.sort_values("Percentage_sorted",ascending=False)
```

```python
data = pd.read_csv('quize_pandas.zip', compression='zip')
data
```

```python
data.sort_values(["Customer","Product","Model"],inplace=True)
data
```

```python
C_count=data.groupby(["Product","Model"])["Count"].sum()
C_count
```

```python
V_count=data.groupby(["Product","Model"])["Volume"].sum()
V_count
```

```python
merged_data = pd.merge(data, C_count, on=['Product','Model'])
merged_data = pd.merge(merged_data, V_count, on=['Product','Model'])
merged_data.sort_values(["Customer","Product","Model"],inplace=True)
merged_data.rename(columns={merged_data.columns[3]: 'Count_old', merged_data.columns[4]: 'Volume_old',merged_data.columns[5]: 'Count_total', merged_data.columns[6]: 'Volume_total'},inplace=True)
merged_data
```

```python
merged_data.pivot_table(index=['Customer','Product','Model'])
```
