```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## mathplotlib

#### 1 - read company_sales_data.csv with pandas

#### 2 - make a plot with total_profit as x axis and month_number columns as y axis

```python
sales=pd.read_csv('company_sales_data.csv')
sales.head()
```

```python
plt.figure(figsize =(5, 3))
plt.plot(sales['month_number'],sales['total_profit'])
plt.show()
```

#### 3 - add "profit" as x label and "Month number" as y label  "company profit" as title  


#### 4 - each month_number should show in x axis and y axis should be [100000, 200000, 300000, 400000, 500000]

```python
plt.figure(figsize =(5, 3))
plt.plot(sales['month_number'],sales['total_profit'])

plt.ylabel("profit")
plt.xlabel("Month number")
plt.title("company profit")

plt.xticks(np.arange(1, 13,1))
plt.yticks(np.arange(100000, 600000,100000))
```

### style change

#### 5 - make your plot appearance like below(5 parameters should change)

```python
plt.figure(figsize =(5, 3))
plt.plot(sales['month_number'],sales['total_profit'],ls = '--',color = 'r',linewidth='3',marker = 'o',ms = 5,mfc = '#ffff00')

plt.ylabel("profit")
plt.xlabel("Month number")
plt.title("company profit")

plt.xticks(np.arange(1, 13,1))
plt.yticks(np.arange(100000, 600000,100000))
```

### bar plot 
#### 6 - create plot with toothpaste and shampoo column (if you can make background girdy)
#### use width and align='edge' to print bars beside each others not top of each others)

```python
barWidth = 0.25
r1=np.arange(1,13)

plt.bar(r1, sales['shampoo'], color='darkorange', width=barWidth, align='edge', label='shampoo sales data')
plt.bar(r1+0.2, sales['toothpaste'], color='steelblue', width=barWidth, align='edge' , label='toothpaste sales data')

plt.grid(linestyle = '--')

plt.xlabel('Month_number')
plt.ylabel('Sales unit in number')
plt.title('toothpaste and shampoo sales data')
plt.xticks(r1)
plt.legend()
plt.show()
```
