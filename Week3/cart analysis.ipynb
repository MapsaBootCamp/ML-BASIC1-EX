```python
sdate = pd.to_datetime('2020-1-1')
edate = pd.to_datetime('2020-12-31')
df['Date'] = pd.to_datetime(df['Date'])
filtered_df = df[(df['Date'] >= sdate) & (df['Date'] <= edate)].sort_values(by="Date")
filtered_df.pivot_table(index=['Customer Id'], aggfunc="count").sort_values(by="Product", ascending=False).index[:5]
```
