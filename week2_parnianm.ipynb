```python
def cast_count(ns, l):
  dictionary = {}
  for n in ns:
    count = 0
    for col in l:
      for i in range(len(col)):
        if n == col[i]:
          count += 1
        else:
          pass
      dictionary[str(n)] = count
  return dictionary



cast_count(names, movies.casts.to_list())

```
