# Question1

با استفاده از دادگان ارائه شده محاسبه کنید هر مشتری چند درصد از هر محصول را خریده است. (با دقت دهگان درصد)

```python
def column_mapping_function(column):
    return column.map(lambda item: round(item/column.sum() * 100, -1) if column.sum() != 0 else None)
```

```python
cpm_p_table = pd.pivot_table(data, index="Customer", columns=["Product", "Model"], values=["Count", "Volume"], aggfunc={"Count": "sum", "Volume": "sum"}).apply(column_mapping_function).dropna(axis = 1, how = 'all')
```

# Question2

به هر سطر لیست دادگان میزان کل محصول خریداری شده اضافه کنید و بنا به شماره مشتری، محصول و مدل مرتب کنید

```python
def group_mapping_function(group):
    group["Count_total"] = group.Count.agg("sum")
    group["Volume_total"] = group.Volume.agg("sum")
    return group
```

```python
ordered_list_sorted = data.groupby(["Product", "Model"], group_keys=False).apply(group_mapping_function).sort_values(["Customer", "Product", "Model"])
ordered_list_sorted = ordered_list_sorted.set_index(["Customer", "Product", "Model"])
```

# Question3

دیتای زیر را برای افراد ۱۸ سال و به بالا بر اساس ستون درصدمرتب کرده تا به شکل نهایی نشان داده شده برسد

```python
import pandas as pd

record = {

'Name': ['Ankit', 'Amit', 'Aishwarya', 'Priyanka', 'Priya', 'Shaurya' ],
'Age': [21, 19, 20, 18, 17, 21],
'Stream': ['Math', 'Commerce', 'Science', 'Math', 'Math', 'Science'],
'Percentage': [88, 92, 95, 70, 65, 78] }

# create a dataframe
dataframe = pd.DataFrame(record, columns = ['Name', 'Age', 'Stream', 'Percentage'])

dataframe
```

```python
df = dataframe[dataframe.Age >= 18].sort_values("Percentage", ascending=False, ignore_index=True)
df.rename(columns={"Name": "Firstname", "Percentage": "Percentage_sorted"}, index= lambda index: chr(ord('A') + index), inplace=True)
```
