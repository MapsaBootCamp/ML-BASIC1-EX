# mathplotlib

#### 1 - read company_sales_data.csv with pandas

#### 2 - make a plot with total_profit as x axis and month_number columns as y axis

```python
company_sales_df = pd.read_csv("Data/company_sales_data.csv")
```

```python
plt.figure(figsize=(4, 3))
plt.plot("month_number", "total_profit", data=company_sales_df, alpha=0.8)
```

#### 3 - add "profit" as x label and "Month number" as y label  "company profit" as title  


#### 4 - each month_number should show in x axis and y axis should be [100000, 200000, 300000, 400000, 500000]

```python
plt.figure(figsize=(4, 3))
plt.plot("month_number", "total_profit", data=company_sales_df, alpha=0.8)
plt.xlabel("Month number")
plt.ylabel("profit")
plt.title("company profit")
plt.xticks(company_sales_df["month_number"])
plt.yticks(range(100000, 600000, 100000))
plt.show()
```

#### style change

#### 5 - make your plot appearance like below(5 parameters should change)

```python
plt.figure(figsize=(4, 3))
plt.plot("month_number", "total_profit", data=company_sales_df, color="red", linewidth = 3, linestyle="--", marker = "o", markerfacecolor="yellow", markersize=4.5)
plt.xlabel("Month number")
plt.ylabel("profit")
plt.title("company profit")
plt.xticks(company_sales_df["month_number"])
plt.yticks(range(100000, 600000, 100000))
plt.show()
```

#### bar plot 
#### 6 - create plot with toothpaste and shampoo column (if you can make background girdy)
#### use width and align='edge' to print bars beside each others not top of each others)

```python
plt.figure(figsize=(4, 3))
plt.bar("month_number", "toothpaste", data=company_sales_df, align="edge", label="Shampoo sales data", width=0.3, )
plt.bar("month_number", "shampoo", data=company_sales_df, align="edge", label="Toothpaste sales data", width=-0.3)
plt.xlabel("Month number")
plt.ylabel("Sales units in number")
plt.title("Toothpaste and shampoo sales data")
plt.xticks(company_sales_df["month_number"])
plt.yticks(range(1000, 9000, 1000))
plt.grid(linestyle = '--', linewidth = 0.7)
plt.legend()
plt.show()
```

# seaborn 

#### 9 - load titanic dataset with seaborn  and plot the count of survivors according to sex

```python
titanic = sns.load_dataset("titanic")
```

```python
plt.figure(figsize=(4, 3))
sns.countplot(x="survived", hue="sex", data=titanic, palette="magma")
plt.show()
```

####  10 - plot the below plot from titanic dataset

```python
plt.figure(figsize=(4, 3))
sns.barplot(x="pclass", y="survived", data=titanic, orient="h", palette="PuRd")
plt.show()
```

### 10 - 
use seaborn and make plot with dowjones dataet like below

```python
dowjones = sns.load_dataset("dowjones")
```

```python
sns.relplot(data=dowjones, x="Date", y="Price", kind="line", height=3.5)
plt.xticks(dowjones["Date"])
plt.show()
```

#### what function can we use instead of kind="line" in seaborn? and the output with pervois one are the same

```python
plt.figure(figsize=(4, 3))
sns.lineplot(data=dowjones, x="Date", y="Price")
plt.xticks(dowjones["Date"])
plt.show()
```
