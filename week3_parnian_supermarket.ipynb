```python
from pandas.core.algorithms import unique
import pandas as pd
df_basket = pd.read_csv("/content/supermarket.csv")
df_basket.head()
```

تعداد محصولات یکتا در کل دادگان، چندتاست؟


```python
def unique_val_count(df, col):
  uni_val = pd.unique(df[str(col)])
  uni_val_list = uni_val.tolist()
  count = len(uni_val_list)
  return count

print(" number of unique values in Customer Id : ", unique_val_count(df_basket, 'Customer Id'))
print(" number of unique values in Date : ", unique_val_count(df_basket, 'Date'))
print(" number of unique values in Product : ", unique_val_count(df_basket, 'Product'))
```

میزان فروش به ازای هر روز به صورت میانگین چند است (تا دو رقم بعد از اعشار)؟


```python
def unique_val(df, col):
  uni_val = pd.unique(df[str(col)])
  uni_val_list = uni_val.tolist()
  return uni_val_list
```

```python
def get_count_filter(df, col, val):
  df1 = df.where(df[str(col)] == val)
  df1 = df1.dropna()
  return df1
```

```python
list_daily_sell = {}
for day in unique_val(df_basket, "Date"):
  per_day_count = get_count_filter(df_basket, "Date", day)
  list_daily_sell[str(day)] = per_day_count.shape[0]

print(max(list_daily_sell, key=list_daily_sell.get), ":", max(list_daily_sell.values()))
```

چهار محصولی که کمتر از بقیه در سبد مشتریان قرار گرفته‌اند، کدامند (ترتیب مهم نیست)؟


```python
list_product_sell = {}
for prod in unique_val(df_basket, "Product"):
  per_prod_count = get_count_filter(df_basket, "Product", prod)
  list_product_sell[str(prod)] = per_prod_count.shape[0]
```

```python
dict_val = list_product_sell.values()
sorted_prod = sorted(dict_val)

def get_key(val):

    for key, value in list_product_sell.items():
        if val == value:
            return key

    return "key doesn't exist"

list_4_least_prod = []
for i in sorted_prod[:4]:
  list_4_least_prod.append(get_key(sorted_prod[i]))

list_4_least_prod
```
