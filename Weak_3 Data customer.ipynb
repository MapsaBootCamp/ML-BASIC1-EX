میزان فروش به ازای هر روز به صورت میانگین چند است (تا دو رقم بعد از اعشار)؟

```python
sm = 0
for i in df.Date.unique():
    sm += df[df["Date"]  == i].Product.value_counts().sum()

round(sm / df.Date.nunique() , ndigits=2)
```

```python
str = ""
for i in lst:
    str += i
    if not(i == lst[-1]):
        str += ","

print(str)
```

```python
dic = {}
for j in df[df["Date"] >= "2020-01-01"]["Date"].unique():
    for i in df["Customer Id"].unique():
        if i in dic:
            if (df[(df["Customer Id"] == i) & (df["Date"] == j)].value_counts().sum()) > 0:
                dic[i] += 1
        elif (df[(df["Customer Id"] == i) & (df["Date"] == j)].value_counts().sum()) > 0:
            dic[i] = 1
        
```

```python
dic = {}
for i in df["Customer Id"].unique():
    dic[i] = df[(df["Date"] >= "2020-01-01") & (df["Customer Id"] == i)]["Date"].nunique()
```
