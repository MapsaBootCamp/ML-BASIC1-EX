```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('/home/gemmara/Downloads/quize_pandas.txt')
df.head()
```

## 1


```python
def percent(row, val, df): 
    index = row.index[0]
    model = df.loc[index]['Model']
    product = df.loc[index]['Product']
    total_df = df[(df['Product'] == product) & (df['Model'] == model)]
    total = total_df['Volume'].sum() + total_df['Count'].sum()
    return round((val / total * 10)) * 10 


df.pivot_table(
    index='Customer',
    columns=['Product', 'Model'],
    values=['Count', 'Volume'],
    aggfunc=lambda x: percent(x, x.sum(), df),
    )
```

## 2

```python
df2 = df
df2.rename(columns={'Count': 'Count_old', 'Volume': 'Volume_old'})
grouped = df.groupby(['Product', 'Model'])
df2['Count_total'], df2['Volume_total'] = grouped['Count_old'].transform('sum'), grouped['Volume_old'].transform('sum')
df2.sort_values(by=['Customer', 'Product', 'Model']).set_index(['Customer', 'Product', 'Model'])
```
